```python
import numpy as np
from ast import literal_eval
from matplotlib import pyplot as plt
from shapely.geometry import Polygon
import matplotlib.image as img_
import os


def find_cross(image_path, predict, danger_zone, shape):
    def PolyArea(x, y):  # x  y -   
        return 0.5 * np.abs(np.dot(x, np.roll(y, 1)) - np.dot(y, np.roll(x, 1)))
    
    k1, k2 = shape[0], shape[1]

    Image = img_.imread(image_path)  #  
#     plt.imshow(Image)

    f1 = open(predict)  # 
    f2 = open(danger_zone)  #  

    # --      
    listi = f2.readlines()
    clipping_polygon = ''
    for i in listi:
        clipping_polygon += i
    clipping_polygon = list(clipping_polygon.split(',\n'))
    for i in range(len(clipping_polygon)):
        clipping_polygon[i] = literal_eval(clipping_polygon[i])
    clipping_polygon.append(clipping_polygon[0])

    # --   
    clipped_polygon = []
    subject_polygons = []

    # --   
#     plt.plot(np.array(clipping_polygon)[:, 0], np.array(clipping_polygon)[:, 1])

    # --     
    for j in f1.readlines():
        # --      
        subject_polygon = []
        # --    x  y  
        x, y, width, height = list(j.split()[1:-1])
        subject_polygon = [
            [int(float(x) * k1) - int(float(width) * k1) // 2, int(float(y) * k2) - int(float(height) * k2) // 2],
            [int(float(x) * k1) + int(float(width) * k1) // 2, int(float(y) * k2) - int(float(height) * k2) // 2],
            [int(float(x) * k1) + int(float(width) * k1) // 2,
             int(float(y) * k2) + int(float(height) * k2) // 2],
            [int(float(x) * k1) - int(float(width) * k1) // 2, int(float(y) * k2) + int(float(height) * k2) // 2]]
        subject_polygon.append(subject_polygon[0])
        subject_polygon = np.array(subject_polygon)
        subject_polygons.append(subject_polygon)
        # --  
#         plt.plot(subject_polygon[:, 0], subject_polygon[:, 1], color='g')
        # --      
        Inter = Polygon(subject_polygon).intersection(Polygon(clipping_polygon))
        # --   
        clipped_polygon.append(Inter)

    # --          
    result = []
    for i in range(len(clipped_polygon)):
        if clipped_polygon[i].geom_type == 'Polygon':
            x, y = clipped_polygon[i].exterior.xy
            result.append(round(PolyArea(x, y) / PolyArea(subject_polygons[i][:, 0], subject_polygons[i][:, 1]), 4) * 100)
#             plt.plot(x, y, color='r')
#             print(round(PolyArea(x, y) / PolyArea(subject_polygons[i][:, 0], subject_polygons[i][:, 1]), 4) * 100, '%',
#                   sep='')
        elif clipped_polygon[i].geom_type == 'MultiPolygon':
            for z in range(len(list(clipped_polygon[i].geoms))):
                x, y = clipped_polygon[i].geoms[z].exterior.xy
                result.append(round(PolyArea(x, y) / PolyArea(subject_polygons[i][:, 0], subject_polygons[i][:, 1]), 4) * 100)

#                 plt.plot(x, y, color='r')
#                 print(round(PolyArea(x, y) / PolyArea(subject_polygons[i][:, 0], subject_polygons[i][:, 1]), 4) * 100,
#                       '%',
#                       sep='')
    return result
```

```python
import csv


def find_cam(filename):
    for i in range(len(cameras)):
        if filename in cameras_imgs[i]:
            return cameras[i]
    
path = r'test_dataset_test/videos'
data_path = r'project\\results_detect'
path_dz = r'project\\danger_zones' 

cameras = os.listdir(path)
cameras.append('Spp-K1-1-2-6_zone1')
cameras.append('Spp-K1-1-2-6_zone2')
cameras.append('Spp-K1-1-2-6_zone3')
cameras.append('Spp-K1-1-2-6_zone4')
cameras_imgs = [os.listdir(os.path.join(path, cam)) for cam in cameras[:-4]]
d = {
    'DpR-Csp-uipv-ShV-V1':(1920, 1080),
    'Pgp-com2-K-1-0-9-36':(1280, 720),
    'Pgp-lpc2-K-0-1-38':(1280, 720),
    'Phl-com3-Shv2-9-K34':(1920, 1080),
    'Php-Angc-K3-1':(1920, 1080),
    'Php-Angc-K3-8':(1920, 1080),
    'Php-Ctm-K-1-12-56':(1920, 1080),
    'Php-Ctm-Shv1-2-K3':(1920, 1080),
    'Php-nta4-shv016309-k2-1-7':(1920, 1080),
    'Spp-210-K1-3-3-5':(1280, 720),
    'Spp-210-K1-3-3-6':(1280, 720),
    'Spp-K1-1-2-6':(1280, 720)
    }

# img_names = [name for name in names if name[-3:] != 'txt']
# print(len(names), len(img_names))
with open('detect_predict.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(['camera_name',"frame_filename","in_danger_zone","percent"])
    for cam in cameras[:-4]:
        cam_images = os.listdir(os.path.join(path, cam))
        
        for img in cam_images:
            curr_cam = find_cam(img)
            if not os.path.isfile(os.path.join(data_path, img[:-3] + 'txt')):
                continue
            if 'danger_' + curr_cam + '.txt' == 'danger_Spp-K1-1-2-6.txt':
                continue
            percent = find_cross(image_path=os.path.join(data_path, img), 
                                 predict=os.path.join(data_path, img[:-3] + 'txt'), 
                                 danger_zone=os.path.join(path_dz, 'danger_' + curr_cam + '.txt'), 
                                 shape=d[curr_cam])
            if percent == 'None':
                continue
            for i, p in enumerate(percent):
                in_danger_zone = True if percent[i] > 15 else False
                writer.writerow([curr_cam, img, in_danger_zone, percent[i]/100])
            
            curr_cam =''
```
